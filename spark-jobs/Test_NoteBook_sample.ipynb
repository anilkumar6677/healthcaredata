```python
from pyspark.sql.session import SparkSession

spark = SparkSession.builder.getOrCreate()

data = [
    (1, "Alice", "HR", 50000),
    (2, "Bob", "IT", 60000),
    (3, "Suresh", "HR", 55000),
    (4, "Anil", "IT", None),
    (5, None, "Finance", 65000)
]
columns = ["emp_id", "name", "department", "salary"]
df = spark.createDataFrame(data, columns)
df.show()
```
